```python
!pip install openai
import openai

# 발급받은 API 키 설정
OPENAI_API_KEY = "sk-proj-sWmkpDnGMRCqpn2t6gJVT3BlbkFJYpIJjLOIcnaBSf3BRxrM"
# openai API 키 인증
#openai.api_key = OPENAI_API_KEY

input_prompt = '''
You are a classifier of external tools, Your job is to indicate each tool belongs to which categories by adding prefix.

Tools are following:
Google[input]: Worker that searches results from Google. Useful when you need to find short and succinct answers about a specific topic. Input should be a search query.
Wikipedia[input]: Worker that search for similar page contents from Wikipedia. Useful when you need to get holistic knowledge about people, places, companies, historical events, or other subjects. The response are long and might contain some irrelevant information. Input should be a search query.
WolframAlpha[input]: A WolframAlpha search engine. Useful when you need to solve a complicated Mathematical or Algebraic equation. Input should be an equation or function.
Calculator[input]: A calculator that can compute arithmetic expressions. Useful when you need to perform math calculations. Input should be a mathematical expression
LLM[input] : A pretrained LLM like yourself. Useful when you need to act with general world knowledge and common sense. Prioritize it when you are confident in solving the problem yourself. Input can be any instruction.

There are two prefix according to their categories of [#1. Task Type] and [#2. Domain Relevance].
[#1. Information Retrieval] [#2. Common Domain] : Task type is to retrieve information and provide that to user. Domain is not limited.
[#1. Task Completion] [#2. Math and Science] : Task type is to do a specific action and return the result to user. Domain is mathmatics and science.

Output should be like
Example:
[#1. Information Retrieval] [#2. Common Domain] Google[input]: Worker that searches results from Google. Useful when you need to find short and succinct answers about a specific topic. Input should be a search query.

'''

prompt = '''
Now, classify all the tools by adding prefix to each tools

'''

def ChatGPT(input_prompt, prompt, API_KEY=OPENAI_API_KEY):
    # api key 세팅
    openai.api_key = API_KEY

    # ChatGPT API 호출 및 최신 언어 모델인 text-davinci-003을 가져옴
    completion = openai.chat.completions.create(
        model='gpt-4-0613',  # 'text-curie-001'  # 'text-babbage-001' #'text-ada-001'
        messages=[
            {"role": "system", "content": input_prompt},
            {"role": "user", "content": prompt},
        ],
        temperature=0)

    return completion.choices[0].message.content


output = ChatGPT(input_prompt, prompt)
print(output)






```
